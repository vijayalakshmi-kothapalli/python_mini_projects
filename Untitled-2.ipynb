```python
a1=[]
a2=[]
a3=[]
a4=[]
a5=[]

```

```python
import random
for _ in range(1000):
   a1.append(random.randint(0,100))
   a2.append(random.randint(50,70))
   a3.append(random.randint(10,20))
   a4.append(random.randint(20,30))
   a5.append(random.randint(40,60))
```
